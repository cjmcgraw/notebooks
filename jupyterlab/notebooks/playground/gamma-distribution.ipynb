```python
import scipy.stats
import numpy as np
import pymc as pm

from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
samples = np.random.gamma(1000, 1/200, size=100_000)
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.hist(samples, bins=50, alpha=0.6)
ax.set_title('np.random.gamma(1000, 1/200)')
```

```python
samples = scipy.stats.gamma(1000, scale=1/200).rvs(100_000)
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.hist(samples, bins=50, alpha=0.6)
ax.set_title('np.random.gamma(1000, 1/200)')
```
